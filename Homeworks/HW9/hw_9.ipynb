Я взяла datase из Kaggle и вытащила данные только по 2 эмоциям и оставила только 50 строк, чтобы можно было с ними работать

```python
data = []
with open('train.txt', 'r') as file:
    for line in file.readlines():
        text, emotion = line.strip().split(';')
        if emotion == 'joy' or emotion == 'sadness':
            data.append((text, emotion))

print(len(data))
    
```

```python
from random import sample, seed


seed(41)
data_sample = sample(data, 50)
```

```python
data_sample
```

```python
import pandas as pd

df = pd.DataFrame(data_sample, columns=['text','emotion'])
y = df['emotion']
df.drop(['emotion'], axis=1, inplace=True)
df.head()
```

```python
y
```

```python
positive_words = ['happy', 'excited', 'awesome', 'love', 'like', 'proud', 'laugh', 'energetic', 'thankful', 'artistic', 'passionate', 'playful', 'amazing', 
                  'joyful', 'generous', 'honesty', 'wonderful', 'welcomed', 'honored','triumphant']
negative_words = ['disappointed', 'frustrated', 'terrible', 'boring' ,'lonely', 'stupid', 'unwanted', 'unwelcome', 'shamed', 'disheartened', 'terrific', 'alone',
                  'neglectful','lazy', 'worthless', 'disillusioned','broke', 'homesick', 'ungrateful']
```

```python
def get_sentiment(text):
    
    num_positive = sum([1 for word in text.split(' ') if word in positive_words])
    num_negative = sum([1 for word in text.split(' ') if word in negative_words])
    # Присвоение метки на основе чистой оценки настроения
    if num_positive > num_negative:
        return 'joy'
    elif num_negative > num_positive:
        return 'sadeness'
```

```python
df['sentiment'] = df['text'].apply(get_sentiment)
df['sentiment'].value_counts()
```

```python
df.to_csv('sentiment.csv', index=True, header=True, encoding='utf-8')
df.head()
```

```python
data_json = pd.read_json('project-6-at-2024-08-08-19-58-a7005909.json')
```

```python
result = []
for index, row in data_json.iterrows():
    choice = row['annotations'][0]['result'][0]['value']['choices'][0]
    ind = row['data']['index']
    result.append((ind, choice))
```

```python
df_from_label = pd.DataFrame(result, columns=['index', 'emotion'])
df_from_label.set_index('index', inplace=True)
df_from_label.head()
```

```python
combined_df = pd.concat([df, df_from_label], axis=1)
combined_df.head()
```

```python
combined_df['sentiment'].fillna(df_from_label['emotion'], inplace=True)
combined_df
```

```python
combined_df.drop(['emotion'], axis=1, inplace=True)
combined_df
```

```python
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x = combined_df['text']
y = combined_df['sentiment']

vectorizer = TfidfVectorizer()
x_vec = vectorizer.fit_transform(x)
X_train, X_test, y_train, y_test = train_test_split(x_vec, y, test_size=0.2, random_state=42)
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
accuracy
```
